## Data Analysis

Important statistical questions are analyzed to understand student performance trends.

```python
average_grade = df['G3'].mean()

print("Average Final Grade:", average_grade)
```

```python
average_grade = df['G3'].mean()

print("Average Final Grade:", average_grade)
```

```python
df = pd.read_csv("student-mat.csv", delimiter=';')

df.head()
```

```python
average_grade = df['G3'].mean()

print("Average Final Grade:", average_grade)
```

```python
high_scores = df[df['G3'] > 15]

print("Students scoring above 15:", high_scores.shape[0])
```

```python
correlation = df['studytime'].corr(df['G3'])

print("Correlation:", correlation)
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

print(gender_avg)
```

## Data Visualization

Graphs are created to visually understand grade distribution and relationships between variables.

```python
plt.figure(figsize=(8,5))

plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')

plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['studytime'], df['G3'], color='green')

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

```python
plt.figure(figsize=(6,5))

gender_avg.plot(kind='bar')

plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
```
